# Accuracy without GA

```python
start_time = time.time()
model1 = MLPClassifier()
Aa1 = 0
Ss1 = 0
Sf1 =0
kfold =10

kf = KFold(n_splits=kfold)
kf.get_n_splits(X)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    
        
    model1.fit(X_train, Y_train.ravel())
    PL1=model1.predict(X_test)
    AC1=model1.score(X_test,Y_test)
    Y_pred = PL1.tolist()
    SS1 = metrics.recall_score(Y_test, Y_pred, pos_label='Normal')
    SF1 = metrics.recall_score(Y_test, Y_pred, pos_label='Cad')

    Aa1 += AC1
    Ss1 += SS1
    Sf1 += SF1

print()
print("Accuracy for MLP without GA: %f" %(Aa1/kfold))
print("Sensitivity for MLP without GA: %f" %(Ss1/kfold))
print("Specificity for MLP without GA: %f" %(Sf1/kfold))
print("--- %s seconds ---" % (time.time() - start_time))
```
